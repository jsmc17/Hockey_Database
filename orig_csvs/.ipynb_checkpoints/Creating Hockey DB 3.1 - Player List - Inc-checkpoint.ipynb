Long time running, so split up the loop into several pieces, then concat together

```python
import string

pt1 = string.ascii_lowercase[:2]
pt1_1 = string.ascii_lowercase[2:5]
pt2= string.ascii_lowercase[5:10]
pt3 = string.ascii_lowercase[10:12]
pt3_1 = string.ascii_lowercase[12:15]
pt4 = string.ascii_lowercase[15:20]
pt5 = string.ascii_lowercase[20:]

print(pt1, pt1_1,pt2,pt3,pt4,pt5)

pt3, pt3_1
```

```python
pid = []
names = []
heights = []
weights = []
birth_dates = []
birth_years = []
cities = []
states = []
country_codes = []

for l in pt4:
    
    if l != 'x':
        all_url = f'https://www.hockey-reference.com/players/{l}/'
        print(all_url)
        resp = requests.get(all_url)
        soup = BeautifulSoup(resp.text, 'html.parser')
        slinks = soup.select("#div_players a")
        surl = ['https://www.hockey-reference.com'+s['href'] for s in slinks]
        
        
        
        
        
        for i in range(len(surl)):
            if i % 37 == 0:
                time.sleep(3)
            resp2= requests.get(surl[i])
            soup2 = BeautifulSoup(resp2.text, 'html.parser')
            slinks2 = soup2.select("#meta p")


            pl_id = slinks[i].attrs['href'].split('/')[3].split('.')[0]
            pl_nm = soup2.select('span')[8].text
            pl_inf = [slinks2[y].text for y in range(len(slinks2))]


            



            if 'Left' in pl_inf[0]:
                shoots = 'Left'
            elif 'Right' in pl_inf[0]:
                shoots = 'Right'
            else:
                shoots = ''

            if 'cm' in pl_inf[1]:
                height = pl_inf[1].split(',')[0]
            else:
                height = ''
            if 'lb' in pl_inf[1]:
                if len(pl_inf[1].split(',')) == 1:
                    weight = pl_inf[1][0:3]
                else:
                    weight = pl_inf[1].split(',')[1][1:4]
            else:
                weight = ''

            if 'Born' in pl_inf[1]:
                bd = pl_inf[1].split('Born: ')[1].split(',')[0]
                byr = pl_inf[1].split('Born: ')[1].split(',')[1][1:5]
            elif len(pl_inf) > 2:
                if 'Born' in pl_inf[2]:
                    bd = pl_inf[2].split('Born: ')[1].split(',')[0]
                    byr = pl_inf[2].split('Born: ')[1].split(',')[1][1:5]
            else:
                bd = ''
                byr = ''


            if ' in' in pl_inf[1]:
                locat = pl_inf[1].split(' in')[1]
                locat_norm = normalize('NFKD', locat).strip()
                if len(locat_norm.split(',')) == 1:
                    country_code = locat_norm.split('  ')[1].strip()
                    city = locat_norm.split(',')[0]
                    state = ''
                else:
                    state = locat_norm.split(',')[1].strip().split('  ')[0]
                    country_code = locat_norm.split(',')[1].strip().split('  ')[-1]
                    city = locat_norm.split(',')[0]
            
            if len(pl_inf) > 2:
                if ' in' in pl_inf[2]:
                    locat = pl_inf[2].split(' in')[1]
                    locat_norm = normalize('NFKD', locat).strip()
                    if len(locat_norm.split(',')) == 1:
                        country_code = locat_norm.split('  ')[1].strip()
                        city = locat_norm.split(',')[0]
                        state = ''
                    else:
                        state = locat_norm.split(',')[1].strip().split('  ')[0]
                        country_code = locat_norm.split(',')[1].strip().split('  ')[-1]
                        city = locat_norm.split(',')[0]
                else: 
                    city = ''
                    state = ''
                    country_code = ''
            else: 
                    city = ''
                    state = ''
                    country_code = ''

            print(pl_id, pl_nm, height, weight, bd, byr, city, state, country_code)
            pid.append(pl_id)
            names.append(pl_nm)
            heights.append(height)
            weights.append(weight)
            birth_dates.append(bd)
            birth_years.append(byr)
            cities.append(city)
            states.append(state)
            country_codes.append(country_code)
```

```python
df1 = pd.read_csv('Plista-b.csv')
df2 = pd.read_csv('Plistc-e.csv')
df3 = pd.read_csv('Plistf-j.csv')
df4 = pd.read_csv('Plistk-l.csv')
df5 = pd.read_csv('Plistm-o.csv')
df6 = pd.read_csv('Plistp-t.csv')
df7 = pd.read_csv('Plistw-z.csv')
```

```python
df = pd.concat([df1,df2,df3,df4,df5,df6,df7])
```

```python
inches = []
lhght = list(df['height'])
lspl = list(df['height'].str.split('-'))
for i in range(len(df['height'])):
    if type(lhght[i]) != float:
        try:
            ht = (int(lspl[i][0]) * 12 + int(lspl[i][1]))
        except ValueError:
            ht = ''
    else:
        ht = ''
    inches.append(ht)
    

dates = []
ldob = list(df['dob'])
lyob = list(df['yob'])
for i in range(len(df['dob'])):
    if (type(ldob[i]) != float):
        try:
            dt = (ldob[i] + ' ' + str(int(lyob[i]))).lstrip()
        except ValueError:
            dt = ''
    else:
        dt = ''
    dates.append(dt)
    
dts = []
for i in range(len(df['dob'])):
    if dates[i] != '':
        dtt = datetime.datetime.strptime(dates[i], '%B %d %Y') 
    else:
        dtt = ''
    dts.append(dtt)
    
    
df['height_in'] = inches
df['born'] = dts
```

```python
df.to_csv('Player_List.csv', index = False)
```
