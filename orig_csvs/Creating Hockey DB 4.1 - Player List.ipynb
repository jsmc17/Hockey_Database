# Transform

## All Transform

```python
poss = []
bds = []
shoots_list = []
height_list_imp = []
height_list_met = []
weights = []
cities = []
states = []
countries = []

for i in range(len(df1['info'])):
    strang = df1['info'][i]
    
    # Position
    if 'Position: ' in strang:
        mo = re.search('Position: \w*([/]\w*)?', strang)
        mo_str = mo.group().split('Position: ')[1]
        poss.append(mo_str)
    else:
        poss.append('')

    # Date of Birth    
    if re.search(r'Born:  \w* \d{1,2},\\xa\d*', strang):
        mo = re.search(r'Born:  \w* \d{1,2},\\xa\d*', strang)
        mo_str = mo.group()
        byr = mo_str[-4:]
        bd = mo_str.split('Born:  ')[1].split(',')[0]
        b_str = bd + ' ' + byr
        b_dt = datetime.datetime.strptime(b_str, '%B %d %Y')
        b_strf = datetime.datetime.strftime(b_dt, '%Y-%m-%d')  #getting proper sql date formatting
        bds.append(b_strf)
    else:
        bds.append('')
        
    # Shoots
    if 'Shoots: ' in strang:
        mo = re.search(r'Shoots: \w*', strang)
        mo_str = mo.group()
        shoots = mo_str.split('Shoots: ')[1]
        shoots_list.append(shoots)
    else:
        shoots_list.append('')
        
    # Height    
    if re.search('\d-\d{1,2}', strang):
        mo = re.search('\d-\d{1,2}', strang)
        mo_str = mo.group()
        height_list_imp.append(mo_str)
    else:
        height_list_imp.append('')
        
    if re.search('\d*cm', strang):
        mo = re.search('\d*cm', strang)
        mo_str = mo.group() 
        height_cm = mo_str.split('cm')[0]
        height_list_met.append(height_cm)
    else:
        height_list_met.append('')
        
    # Weight
    if re.search('\d*lb', strang):
        mo = re.search('\d*lb', strang)
        mo_str = mo.group()
        wt = mo_str.split('lb')[0][1:]
        weights.append(wt)
    else:
        weights.append('')
        
    # Birthplace
    if re.search(r'in\\xa0\w*,\\xa0\w*  \w*', strang):
        mo = re.search(r'in\\xa0\w*,\\xa0\w*  \w*', strang)
        mo_str = mo.group()
        city = mo_str.split('in\\xa0')[1].split(',')[0]
        state = mo_str.split('in\\xa0')[1].split(',\\xa0')[1].split(' ')[0]
        country = mo_str.split('in\\xa0')[1].split(',\\xa0')[1].split('  ')[1]
        cities.append(city)
        states.append(state)
        countries.append(country)
    else:
        cities.append('')
        states.append('')
        countries.append('')

df1['Position'] = poss
df1['dob'] = bds
df1['Shoots'] = shoots_list
df1['height'] = height_list_imp
df1['height_cm'] = height_list_met
df1['weight']  = weights
df1['city'] = cities
df1['state'] = states
df1['country'] = countries
```
